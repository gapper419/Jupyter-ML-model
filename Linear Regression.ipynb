# **Data preparation** 
### is a crucial step in machine learning and data modeling because raw data is often messy,
### inconsistent, and incomplete. Preparing data properly ensures better model performance,
### accuracy, and generalizability. Here are the key reasons why we prepare data:  -

## creating a column 'y' variable for preparation of data modelling

```python

y = df.iloc[: ,8]
#y.shape
#.info()
y.head()
```

## creating a column 'X' variable for preparation of data modelling

```python
X = df.iloc[ : , [3,4,6,7,9,10,15]]
X
```

## onehotencoder converting categorical variables into numerical format

```python

X = pd.get_dummies(X , drop_first= True)
X.head(5)
```

## Split dataset into training and testing sets

```python

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

## standarzing the dataset

```python

scaler = StandardScaler()
scaler.fit(X)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## creating the data model 

```python

mdl = LogisticRegression(max_iter=1000)
```

## fitting the dataset for training

```python

mdl.fit(X_train , y_train)
```

## probablity score

```python

y_prob = mdl.predict_proba(X_test)
y_prob[0:10]
```

## the predicted variables in 'y' dataset

```python

y_pred = mdl.predict(X_test)
y_pred
```

## Model evaluation

```python

accuracy_score(y_test, y_pred)
#0.7379949452401011
```

# #Evaluation metrics

```python

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
```

## Model evaluation

```python

report = classification_report(y_test , y_pred )
print('Classification Report:\n', report )
```

# **Metrics Explained**

## **R² Score (r2_score)**
    -The R² score (coefficient of determination) measures how well the model explains the variability of the target variable (y)

## **Mean Absolute Error (MAE)**
    -MAE measures the average absolute difference between actual and predicted values. It gives an idea of how much, on average, the model's       
    predictions deviate from the actual values.

## **Root Mean Squared Error (RMSE)**
    -RMSE is the square root of the Mean Squared Error (MSE). It measures the standard deviation of prediction errors. RMSE penalizes larger errors 
    more than MAE due to squaring before averaging.

## **Precision**
    -Measures how many of the predicted positive cases were actually correct
    -High precision means fewer false positives.

## **Recall** 
    -Measures how many of the actual positive cases were correctly predicted.
    -High recall means fewer false negatives.

## **F1-score** 

    -The harmonic mean of precision and recall.
    -A good balance between precision and recall, useful when you need both metrics to be high.

## **Support**

    -The number of actual occurrences of each class in the dataset.
    -Helps understand class distribution and whether the model is biased.
